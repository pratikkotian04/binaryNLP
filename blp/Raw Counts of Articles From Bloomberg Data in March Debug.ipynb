## Use only single day

```python
sub_raw = raw_counts.loc['20100105':'20100105']
```

##Looking at Finance US 

```python
finance_us = sub_raw[['fin_location_US']]
fin_med = raw_counts[['fin_location_US']].median() # Median from entire finance dataset
fin_sub_med = finance_us.median()

fin_plot = finance_us.plot()
fin_plot.axhline(fin_med['fin_location_US'], color=np.random.rand(3,1), linestyle='--',label='Finance Median')
#Debug
#fin_plot.axhline(fin_sub_med['fin_location_US'], color=np.random.rand(3,1), linestyle='--',label='Period Finance Median')
fin_plot.legend()
```
